```python
idx_private_no = college["Private"] == 'No' # Boolean
college_private_no = college[idx_private_no] # Avoir un dataframe sÃ©lectionnÃ©
college_private_no = college_private_no ['Personal']
college_private_no
```

```python
idx_private_yes = college["Private"] == 'Yes' # Boolean
college_private_yes = college[idx_private_yes] # Avoir un dataframe sÃ©lectionnÃ©
college_private_yes = college_private_yes ['Personal']
college_private_yes
```

```python
college_private_no_max = college_private_no.max()
college_private_no_max
```

```python
idx_private_yes_Personal_Exception = college_private_yes > college_private_no_max
idx_private_yes_Personal_Exception.sum()
```
