# For Loops

```python
total = 0

for value in [3, 2, 19] :
    total += value

print('Total is: {0}'.format(total))
```

```python
# The same code
## Code 1
3*(2+3+19) + 2*(2+3+19) + 1*(2+3+19)

# Code 2
for value in [2,3,9] :
    for weight in [3,2,1]:
        total += value +weight
print(total)
```

```python
for value, weight in zip([2,3,19], 
                         [0.2,0.3,0.5]) : 
    total += weight * value
print(total)
```

# String Formatting

```python
rng = np.random.default_rng(1)
A = rng.standard_normal((127,5))
M = rng.choice([0, np.nan], p=[0.8, 0.2], size = A.shape)
A += M
D = pd.DataFrame(A, columns=['food',
 'bar',
 'pickle',
 'snack',
 'popcorn'])
D[:3]
```

```python
for col in D.columns : 
    template = '"{0} has {1:.2%} missing values'
    print(template.format(col,
                          np.isnan(D[col].mean())))
```
